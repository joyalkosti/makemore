```python
def parse_resume(resume_path):
    # Check the file extension to determine the parsing method
    if resume_path.endswith(".docx"):
        # Parse .docx files using docx2txt
        text = docx2txt.process(resume_path)
    else:
        # Parse other file formats using pyresparser
        parser = ResumeParser(resume_path)
        data = parser.get_extracted_data()
        text = data["content"]

    return text
```

```python
import PyPDF2
import spacy
import numpy as np
from transformers import pipeline
import json

def extract_text_from_pdf(file_path):
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        text = ""
        for page in range(num_pages):
            text += reader.pages[page].extract_text()
        return text

nlp = spacy.load("en_core_web_sm")
resume_parser = pipeline("ner", model="bert-base-cased")

resumes_directory = "/content/resume"
parsed_resumes = []

for i in range(1, 6):
    pdf_path = f"{resumes_directory}/resume{i}.pdf"
    resume_text = extract_text_from_pdf(pdf_path)

    parsed_data = resume_parser(resume_text)
    parsed_resumes.append(parsed_data)

def convert_to_json_serializable(obj):
    if isinstance(obj, np.float32):
        return float(obj)
    raise TypeError(f"Object of type {type(obj)} is not JSON serializable")

for parsed_data in parsed_resumes:
    for entity in parsed_data:
        entity["score"] = convert_to_json_serializable(entity["score"])

with open("parsed_resumes1.json", "w") as file:
    json.dump(parsed_resumes, file, default=convert_to_json_serializable)
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
```

```python
def parse_resume_pdf(file_path):
    resource_manager = PDFResourceManager()
    output_string = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    text_converter = TextConverter(resource_manager, output_string, codec=codec, laparams=laparams)
    with open(file_path, 'rb') as file:
        interpreter = PDFPageInterpreter(resource_manager, text_converter)
        for page in PDFPage.get_pages(file):
            interpreter.process_page(page)
        text = output_string.getvalue()
    text_converter.close()
    output_string.close()
    return text
```

```python
nlp = spacy.blank('en')

def get_spacy(file,data):
  nlp=spacy.blank("en")
  db=DocBin()

  for text , annot in tqdm(data):
    doc=nlp.make_doc(text)
    annot=annot['entities']

    ents=[]
    entiy_indices=[]


    for start, end, label in annot:
      skip_entity=False
      for idx in range(start,end):
        if idx in entiy_indices:
          skip_entity=True
          break

      if skip_entity==True:
        continue

      entiy_indices=entiy_indices+ list(range(start,end))


      try:
        span=doc.char_span(start, end, label=label,alignment_mode='strict')

      except:
        continue


      if span is None:
        err_data=(str([start, end])+ "       "+ str(text) + "\n").encode('ascii', 'ignore').decode('ascii')
        # err_data=u' '.join(([start,end], text)).encode('utf-8').strip()
        file.write(err_data)

      else:
        ents.append(span)


    try:
      doc.ents=ents
      db.add(doc)
    except:
      pass

  return db




















```

```python
text = '''
"name": "Abhijit",
"email": "something@gmail.com",
"skills": ["springboot", "react", "web development", "networking"],
"degree": "bachelors of technology",
"language": "english"
'''

# Remove newlines and spaces
text = text.replace('\n', '').replace(' ', '')

# Split the text into key-value pairs
pairs = text.split(',')

# Create an empty dictionary
data = {}

# Loop through the pairs and populate the dictionary
for pair in pairs:
    key, value = pair.split(':')
    data[key.strip('"')] = value.strip('"')

# Print the resulting dictionary
for key, value in data.items():
    if key == "skills":
        value = ", ".join([skill.strip('"') for skill in value.split(',')])
    print(f'"{key}": {value}')
```

```python
import fitz
import spacy

def extract_entities_from_pdf(fname, model_path):
    doc = fitz.open(fname)
    nlp = spacy.load(model_path)

    text = " "
    for page in doc:
        text = text + str(page.get_text())

    entity_dict = {}
    doc = nlp(text)

    for ent in doc.ents:
        label = ent.label_
        text = ent.text

        if label in entity_dict:
            entity_dict[label].append(text)
        else:
            entity_dict[label] = [text]

    return entity_dict


def main():
    fname = '/content/drive/MyDrive/Colab Notebooks/Resume Pool_AI Team/Sumitha_CV.pdf'
    model_path = '/content/drive/MyDrive/model-last'

    entities = extract_entities_from_pdf(fname, model_path)

    for label, texts in entities.items():
        print(label, ":", texts)


if __name__ == '__main__':
    main()
```
